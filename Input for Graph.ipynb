```python
import dash
import numpy as np
import pandas as pd
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
from dash.dependencies import Input,Output,State
app=dash.Dash()
app.layout=html.Div(
    children=[html.H1('Live Filtring Graph '),
        dcc.Input(id='Input',value='Close',type ='text'),
        dcc.Graph(id='graph'),
        html.Div(['This is OUtput Area'],id='Output')])



@app.callback(
Output(component_id='graph',component_property='figure'),
[Input(component_id='Input',component_property='value')])

def update_value(req):
    df=pd.read_csv('I:\Text Datasets\CSV\yahoo_finance.csv')
    df['Date']=pd.to_datetime(df.Date)       
    return  {            
            'data':[{
                            'x':df['Date'],
                            'y':df[req],                          
                            'type':'marker'
                    
            }],
            'layout':{'title':req,'showlegend':False}
                    
            }
   
if __name__=='__main__':
    app.run_server(port=4050)         
```
